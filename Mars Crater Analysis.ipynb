```python
import pandas
import numpy
import seaborn
import matplotlib.pyplot as plt
```

```python
data = pandas.read_csv('marscrater_pds.csv', low_memory=False)
```

```python
seaborn.distplot(data['LATITUDE_CIRCLE_IMAGE'], kde=False);
plt.xlabel('Latitudes')
plt.title('Latitudes in Mars Crater Data')
```
